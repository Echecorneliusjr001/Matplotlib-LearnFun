```python
#To import libraries and create various plot by plotting x and y axis
import numpy as np
import matplotlib.pyplot as plt
xpoints = np.array ([0, 6])
ypoints = np.array ([0,250])
plt.plot(xpoints, ypoints)
plt.show()
```

```python
#Draw a line in a diagram from position (1, 3) to position (8, 10)
xpoints = np.array ([1, 8])
ypoints = np.array ([3, 10])
plt.plot(xpoints, ypoints)
plt.show()
```

```python
# To Plot without a line
xpoints = np.array ([1, 8])
ypoints = np.array ([3, 10])
plt.plot(xpoints, ypoints, 'o')
plt.show()
```

```python
#TO plot multiple lines, drawing a line in a diagram from position (1, 3) to (2, 8) then to (6, 1) and finally to position (8, 10)
xpoints = np.array([1, 2, 6, 8])
ypoints = np.array([3, 8, 1, 10])
plt.plot(xpoints, ypoints)
plt.show()
```

```python
#Plotting without x axis plots, there will stiil be a default x points
ypoints = np.array ([3, 8, 1, 10, 5, 7])
plt.plot(ypoints)
plt.show()
```

```python
#Using keyword marker to mark each point with a circle 
ypoints = np.array([3, 8, 1, 10])
plt.plot(ypoints, marker = 'o')
plt.show()
```

```python
# to marl each point with a star 
ypoints = np.array([3, 8, 1, 10])
plt.plot(ypoints, marker = '*')
plt.show()
```

```python
# Make each point with a dotted line
ypoints = np.array([3, 8, 1, 10])
plt.plot(ypoints, linestyle = "dotted")
plt.show()
```

```python
# to make a circle at the point of the broken line
ypoints = np.array([3, 8, 1, 10])
plt.plot(ypoints, 'o:r')
plt.show()
```

```python
# To make a dashed line 
ypoints = np.array([3, 8, 1, 10])
plt.plot(ypoints, linestyle = "dashed")
plt.show()
```

```python
#to create labels named average pulse and burnage calories for a plot
x = np.array([90, 95, 100, 105, 110, 115, 120, 125, 130, 135])
y = np.array([250, 260, 270, 280, 290, 300, 310, 320, 330, 340])
plt.plot(x, y)
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")
plt.show()
```

```python
# To create a title for a plot 
x = np.array([90, 95, 100, 105, 110, 115, 120, 125, 130, 135])
y = np.array([250, 260, 270, 280, 290, 300, 310, 320, 330, 340])
plt.plot(x, y)
plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")
plt.show()
```

```python
#Settingfont properties for the titles and labels of the plot
x = np.array([90, 95, 100, 105, 110, 115, 120, 125, 130, 135])
y = np.array([250, 260, 270, 280, 290, 300, 310, 320, 330, 340])
plt.plot(x, y)
font1 = {'family':'serif', 'color':'green', 'size':20}
font2 = {'family':'serif', 'color':'blue', 'size':15}
plt.title("Sports Watch Data", fontdict= font1)
plt.xlabel("Average Pulse", fontdict = font2)
plt.ylabel("Calorie Burnage", fontdict = font2)
plt.show()
```

```python
# changing the postition of the title to the left using loc
x = np.array([90, 95, 100, 105, 110, 115, 120, 125, 130, 135])
y = np.array([250, 260, 270, 280, 290, 300, 310, 320, 330, 340])
plt.plot(x, y)
font1 = {'family':'serif', 'color':'green', 'size':20}
font2 = {'family':'serif', 'color':'blue', 'size':15}
plt.title("Sports Watch Data", fontdict= font1, loc ='left')
plt.xlabel("Average Pulse", fontdict = font2)
plt.ylabel("Calorie Burnage", fontdict = font2)
plt.show()
```

```python
#Adding grid function to the plot
x = np.array([90, 95, 100, 105, 110, 115, 120, 125, 130, 135])
y = np.array([250, 260, 270, 280, 290, 300, 310, 320, 330, 340])
plt.plot(x, y)
font1 = {'family':'serif', 'color':'green', 'size':20}
font2 = {'family':'serif', 'color':'blue', 'size':15}
plt.title("Sports Watch Data", fontdict= font1, loc ='left')
plt.xlabel("Average Pulse", fontdict = font2)
plt.ylabel("Calorie Burnage", fontdict = font2)
plt.grid()
plt.show()
```

```python
# Displaying only x axis grid function on the plot 
x = np.array([90, 95, 100, 105, 110, 115, 120, 125, 130, 135])
y = np.array([250, 260, 270, 280, 290, 300, 310, 320, 330, 340])
plt.plot(x, y)
font1 = {'family':'serif', 'color':'green', 'size':20}
font2 = {'family':'serif', 'color':'blue', 'size':15}
plt.title("Sports Watch Data", fontdict= font1, loc ='left')
plt.xlabel("Average Pulse", fontdict = font2)
plt.ylabel("Calorie Burnage", fontdict = font2)
plt.grid(axis = 'x')
plt.show()
```

```python
#Setting  grid line properties for both x and y axis on the plot
x = np.array([90, 95, 100, 105, 110, 115, 120, 125, 130, 135])
y = np.array([250, 260, 270, 280, 290, 300, 310, 320, 330, 340])
plt.plot(x, y)
font1 = {'family':'serif', 'color':'green', 'size':20}
font2 = {'family':'serif', 'color':'blue', 'size':15}
plt.title("Sports Watch Data", fontdict= font1, loc ='left')
plt.xlabel("Average Pulse", fontdict = font2)
plt.ylabel("Calorie Burnage", fontdict = font2)
plt.grid(color = 'green', linestyle = '--', linewidth = 0.5)
plt.show()
```

```python
# Displaying multiple plots (Plots being displayed side by side) in one page by using the subplot() function
#plot 1
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])
plt.subplot(1, 2, 1)# The number represnt the 1 row, 2 columns and the index as the first plot
plt.plot(x,y)

#plot 2
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])
plt.subplot(1, 2, 2)
plt.plot(x,y)

plt.show()
```

```python
#Creating a simple scatter plot showinnd the labels of different axis
x = np.array([5, 7, 8, 7, 2, 17, 2, 9, 4, 11, 12, 9, 6])
y = np.array([99, 86, 87, 88, 111, 86, 103, 87, 94, 78, 77, 85, 86])
plt.ylabel('Speed of the car')
plt.xlabel('Age of the car')
plt.scatter(x, y)
plt.show()
```

```python
#Creating a bar plot
#bar plot 1
x = np.array([5, 7, 8, 7, 2, 17, 2, 9, 4, 11, 12, 9, 6])
y = np.array([99, 86, 87, 88, 111, 86, 103, 87, 94, 78, 77, 85, 86])
plt.bar(x, y, color = 'darkred')

plt.ylabel('Speed of the car')
plt.xlabel('Age of the car')

#Bar plot 2
x = np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
y = np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])
plt.bar(x, y, color= '#88c999')

plt.show()
```

```python
#creating random arrays with 100 values for x points, ypoints, colors and sizes
x = np.random.randint(100, size= (100))
y = np.random.randint(100, size= (100))
colors = np.random.randint(100, size= (100))
sizes =10 * np.random.randint(100, size= (100))
plt.scatter(x, y, s=sizes, c=colors, alpha=0.5, cmap="nipy_spectral")
plt.colorbar()
plt.show()
```

```python
#drawing 4 bars using plt.bar() function, #4CAF50 for green in hexadecimal value
x = np.array(['A', 'B', 'C', 'D'])
y = np.array([3, 7, 1, 10])
plt.bar(x, y, color= 'green')
plt.show()
```

```python
#drawing 4  horizontal bars using plt.barh() function, and using height to get thin bars #4CAF50 for green in hexadecimal value
x = np.array(['A', 'B', 'C', 'D'])
y = np.array([3, 7, 1, 10])
plt.barh(x, y, color= 'green', height = 0.1)
plt.show()
```

```python
#creating a simple histogram plot
x= np.random.normal(170, 10, 250)
plt.hist(x)
plt.show()
```

```python
# drawing piechart using the pie() function
x = np.array([10, 15, 30, 45])
plt.pie(x)
plt.show()
```

```python
# drawing piechart with their labels using the pie() function
x = np.array([10, 15, 30, 45])
mylabels = ['Apples', 'Bananas', 'Cherries', 'Grapes']
plt.pie(x, labels=mylabels)
plt.show()
```

```python
# drawing piechart with start angle of the wedge at 90 using the startangle parameter
x = np.array([10, 15, 30, 45])
mylabels = ['Apples', 'Bananas', 'Cherries', 'Grapes']
plt.pie(x, labels=mylabels, startangle = 90)
plt.show()
```

```python
# drawing piechart to pull the wedge of apple using the explode parameter
x = np.array([10, 15, 30, 45])
mylabels = ['Apples', 'Bananas', 'Cherries', 'Grapes']
myexplode = [0, 0, 0, 0.2]
plt.pie(x, labels=mylabels, explode=myexplode)
plt.show()
```

```python
# drawing piechart and adding shadow to the wedge using the shadow parameter
x = np.array([10, 15, 30, 45])
mylabels = ['Apples', 'Bananas', 'Cherries', 'Grapes']
plt.pie(x, labels=mylabels, shadow = True)
plt.show()
```

```python
# drawing piechart to change the color of each of the wedge using the colors parameter
x = np.array([10, 15, 30, 45])
mylabels = ['Apples', 'Bananas', 'Cherries', 'Grapes']
mycolors= ['red', 'yellow','green', 'purple']
plt.pie(x, labels=mylabels, colors=mycolors)
plt.show()
```

```python
# drawing piechart then adding a list of expalnation for each wedge using the legend function
x = np.array([10, 15, 30, 45])
mylabels = ['Apples', 'Bananas', 'Cherries', 'Grapes']
mycolors= ['red', 'yellow','green', 'purple']
plt.pie(x, labels=mylabels, colors=mycolors)
plt.legend()
plt.show()
```

```python
# drawing piechart then adding a list of expalnation with title for each wedge using the legend function
x = np.array([10, 15, 30, 45])
mylabels = ['Apples', 'Bananas', 'Cherries', 'Grapes']
mycolors= ['red', 'yellow','green', 'purple']
plt.pie(x, labels=mylabels, colors=mycolors)
plt.legend(title ='Fruits')
plt.show()
```
